There are different layer like Conv Layer, Relu Layer and the Pooling Layer that perform feature extraction on an image. 

**Layers in CNN**
1. Convolutional Layer 
2. Pooling Layer 
3. Fully Connected layer(Flatten)


**Image_to_Array** 
In CNN every image represented in the form of an array of pixels. 



**Convolution Layer**
this is the first step to extracting the valuable featurs from an imgae. Conv Layer has a multiple filter to perform a Convolutional Operation. Con Layer perform a task based on that formula; 
     output =  'n-f+1' (here, n is size of an image and f is filter)

**Pooling layer** 
Pooling layer reduces the dimensionality of feature map and it will generate the pooled feature map. 
image--> 
Pooling mainly helps in extracting sharp and smooth features. It is also done to reduce variance and computations. Max-pooling helps in extracting 'low-level features' like edges, points, etc.

After getting the feature map of the convolutional layer will apply max pooling on that feature to exract the more clear image.




**Flatten**
This step is called Flattening. Flattening is used to convert 2 dimensional arrays to single long continuous vector. 
This vector will be input for fully connected layer to classify the image.
